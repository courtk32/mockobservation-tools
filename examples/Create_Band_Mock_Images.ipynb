```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import glob

import math
import matplotlib as mpl

from matplotlib.patches import Ellipse
from mpl_toolkits.axes_grid1 import make_axes_locatable

from mockobservation_tools.general_tools import radius_of_param_limit
from mockobservation_tools.galaxy_tools import load_sim, load_halo, mask_sim_to_halo, get_mock_observation
%matplotlib inline
pd.set_option('display.max_columns', None)
```

```python
# This will be created to have dark matter halo centered images 
# These are made with previous parameters
```

```python
# load all images

image_files_list = glob.glob('../mock_image_files_round2/*.npy')

data = pd.read_csv('../fulldata_allparams.csv')

```

```python
for file in image_files_list:
    image_file = file.split("/")[-1]
    image_params = image_file.split('_')
    image_type = image_params[2]
    #FOV = int(image_params[-2][3:])
    view = image_params[-3]
    sim = image_params[0]+'_'+image_params[1]
    pixel = 1000
    
    #Run through my mock images
    if image_type == 'ugrband':
        print(file)
        if sim in ['m10b_res500', 'm10c_res500', 'm10d_res500', 
                   'm10e_res500', 'm10f_res500', 'm10g_res500',
                   'm10h_res500','m10i_res500', 'm10j_res500',
                   'm10k_res500','m10l_res500','m10m_res500','m10q_res250']:
            snap_num = 184
            filetype = 'ascii'
            halo_path = '/DFS-L/DATA/cosmo/grenache/aalazar/FIRE/GVB/'+str(sim)+'/halo/rockstar_dm/catalog/'
        
        elif sim in ['m11a_res2100','m11b_res2100','m11c_res2100', 
                     'm11d_res7100', 'm11e_res7100','m11h_res7100',
                     'm11i_res7100','m11q_res7100','m11v_res7100']:
            snap_num = 600
            filetype = 'hdf5'
            halo_path = '/DFS-L/DATA/cosmo/grenache/aalazar/FIRE/GVB/'+str(sim)+'/halo/rockstar_dm/hdf5/'
        
        
        sim_data = data.loc[(data['sim'] == sim) & (data['proj'] == view)]
        FOV = int(math.ceil(sim_data['rvir_10']))
        
        halo = load_halo(halo_path, snap_num, host=True, filetype=filetype)

        sim_path =  '/DFS-L/DATA/cosmo/grenache/aalazar/FIRE/GVB/'+str(sim)+'/output/hdf5/snapdir_'+str(snap_num)+'/'
        star_snapdict, gas_snapdict = load_sim(sim_path,snap_num)
                
        star_snapdict, gas_snapdict, halo = mask_sim_to_halo(star_snapdict=star_snapdict,
                                                             gas_snapdict=gas_snapdict, 
                                                             host_halo=halo, 
                                                             orient=False, #if true, then orients toward disk
                                                             lim = True, #Masks particles beyond limvalue  
                                                             limvalue = FOV #Same lim as mock image
                                                            )
        
        
        band_image = get_mock_observation(star_snapdict,gas_snapdict,
                                          FOV=FOV, 
                                          return_type='mock_image',
                                          pixels=pixel, 
                                          view=view,
                                          minden=1, 
                                          dynrange=1e10,
                                          center='none')
        
        save_file_name_mock = '../mass_mock_images/mock_image_halo_centered/' + sim + '_mockimg_' + view + '_FOV' + str(FOV) + '_halocenternosblims.npy'
        save_file_name_band = '../mass_mock_images/mock_image_halo_centered/' + sim + '_ugrband_' + view + '_FOV' + str(FOV) + '_halocenternosblims.npy'
        
        np.save(save_file_name_mock, band_image[0])
        np.save(save_file_name_band, band_image[1:])
        
        plt.imshow(band_image[0])
        plt.gca().axis('equal')
        plt.gca().axis('off')
        plt.show()
        
        
```
